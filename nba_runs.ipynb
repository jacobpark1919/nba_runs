```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/nba.csv')
df
```

```python
df.dtypes
```

```python
# add timeout column: 1 if they called timeout 0 otherwise

df1 = df.copy()
df1['timeout_home'] = df1['descriptionPlayHome'].astype(str).str.contains('timeout', case=False, na=False).astype(int)
df1['timeout_away'] = df1['descriptionPlayVisitor'].astype(str).str.contains('timeout', case=False, na=False).astype(int)
df1
```

```python
# Convert game time to seconds for easier calculations
df1["time_seconds"] = df1["minuteGame"] * 60
```

```python
# Define parameters for a "run"
POINT_DIFFERENCE = 6  # A team must outscore the opponent by at least 6 points
WINDOW_EVENTS = 10  # Consider last 10 events (approx. 2 minutes of game time)

# Compute rolling difference in score for both teams
df1["score_diff_home"] = df1["scoreHome"].diff(WINDOW_EVENTS)
df1["score_diff_away"] = df1["scoreAway"].diff(WINDOW_EVENTS)
```

```python
# Identify runs: If a team outscores the opponent by at least POINT_DIFFERENCE in the rolling window
df1["run_home"] = (df1["score_diff_home"] - df1["score_diff_away"] >= POINT_DIFFERENCE).astype(int)
df1["run_away"] = (df1["score_diff_away"] - df1["score_diff_home"] >= POINT_DIFFERENCE).astype(int)
```
